# Understanding Vectors and Embeddings

A `vector` is an orderded list of numbers that represent an object. In ML, vectors are used to represent words, sentences, images and other data

Imagine a space where we represent human beings, with the X-axis representing weight and the Y-axis representing height.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data: weights (kg) and heights (cm) of 5 individuals as a matrix
data = np.array([
    [60, 160],  # Individual 1: [Weight, Height]
    [72, 175],  # Individual 2
    [85, 180],  # Individual 3
    [90, 190],  # Individual 4
    [55, 155]   # Individual 5
])

# Extracting weights and heights from the matrix
weights = data[:, 0]  # First column: weights
heights = data[:, 1]  # Second column: heights

# Plotting the data
plt.figure(figsize=(8, 6))
plt.scatter(weights, heights, color='blue', label='Individuals')

# Adding labels and title
plt.title('Human Beings Representation on Weight-Height Space', fontsize=14)
plt.xlabel('Weight (kg)', fontsize=12)
plt.ylabel('Height (cm)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
